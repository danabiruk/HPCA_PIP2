# **Fluorescent Profile**


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_cell = pd.read_csv('cell10_cell_ΔF.csv')
red_rois = pd.read_csv('cell10_red_ΔF.csv')
green_rois = pd.read_csv('cell10_green_ΔF.csv')
```

```python
plt.figure(figsize=(6, 5))
plt.plot(all_cell["time"], all_cell["int"], color="darkblue", linewidth=8, alpha=0.8)
plt.plot(red_rois["time"], red_rois["int"], color="red", linewidth=8, alpha=0.8)
plt.plot(green_rois["time"], green_rois["int"], color="green", linewidth=8, alpha=0.8)

plt.xlabel("Time, s", fontsize=50, fontweight='bold')
plt.ylabel("Int., ΔF/F₀", fontsize=50, fontweight='bold')
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.axhline(0, color='black', linewidth=4)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(4)

plt.grid(False)
plt.tight_layout()
plt.subplots_adjust(left=0.15, right=0.95, top=0.95, bottom=0.15)

plt.show()
```
